** Number Guess Game**

```python
import random

def guess_the_number():
    print("Welcome to Guess the Number!")
    number = random.randint(1, 100)
    attempts = 0

    while True:
        try:
            guess = int(input("Guess a number between 1 and 100: "))
            attempts += 1

            if guess < number:
                print("Too low! Try again.")
            elif guess > number:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number in {attempts} attempts.")
                break
        except ValueError:
            print("Please enter a valid number.")

guess_the_number()
```

**Rock, paper, scissors Python Project**

```python
import random

def rock_paper_scissors():
    print("Welcome to Rock, Paper, Scissors! Let's play!!")
    choices = ["rock", "paper", "scissors"]

    while True:
        my_choice = input("Enter rock, paper, or scissors or 'quit' to exit. : ").strip().lower()
        if my_choice == "quit":
            print("Thanks for playing!")
            break

        if my_choice not in choices:
            print("Invalid choice. Please try again.")
            continue

        random = random.choice(choices)
        print(f"Computer chose: {random}")

        outcome = "It's a tie!" if my_choice == random else \
                  "You win!" if (my_choice, random) in [("rock", "scissors"), ("paper", "rock"), ("scissors", "paper")] else "You lose!"
        print(outcome)

rock_paper_scissors()
```

**QR code encoder / decoder Python Project**

```python
!pip install qrcode[pil]
import qrcode
import base64
from io import BytesIO
from IPython.display import HTML, display

def generate_qr(data, filename="qrcode.png"):
    """Generates a QR code for the given data and saves it as a PNG image."""
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)

    img = qr.make_image(fill_color="black", back_color="white")
    img.save(filename)
    print(f"QR code saved as {filename}")

    # Display the QR code in the notebook
    display(img)

    # Create a download link for the QR code
    buffered = BytesIO()
    img.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue()).decode()
    href = f'<a href="data:image/png;base64,{img_str}" download="{filename}">Download QR Code</a>'
    display(HTML(href))


if __name__ == "__main__":
    data_to_encode = input("Enter the text or URL to encode: ")
    generate_qr(data_to_encode)
```

**Basic Command-Line Tic-Tac-Toe**

```python
def print_board(b):
    for row in b:
        print(" | ".join(row)); print("-" * 9)

def check_win(b, p):
    for r in b:
        if all(c == p for c in r): return True
    for c in range(3):
        if all(b[i][c] == p for i in range(3)): return True
    return all(b[i][i] == p for i in range(3)) or all(b[i][2 - i] == p for i in range(3))

def is_full(b):
    return all(c != " " for row in b for c in row)

def play_game():
    b = [[" "] * 3 for _ in range(3)]
    players = ["X", "O"]; i = 0
    while True:
        p = players[i]; print_board(b); print(f"Player {p}'s turn.")
        while True:
            try:
                r, c = map(int, input("Enter row, col (0-2): ").split(','))
                if 0 <= r < 3 and 0 <= c < 3 and b[r][c] == " ":
                    b[r][c] = p; break
                else: print("Invalid move.")
            except ValueError: print("Invalid input.")
        if check_win(b, p): print_board(b); print(f"Player {p} wins!"); break
        if is_full(b): print_board(b); print("It's a draw!"); break
        i = 1 - i

if __name__ == "__main__":
    play_game()
```

**Binary Search Python Project**

```python
def Binary_search(arr, t):
    l, r = 0, len(arr) - 1
    while l <= r:
        m = (l + r) // 2
        if arr[m] == t: return m
        elif arr[m] < t: l = m + 1
        else: r = m - 1
    return -1

if __name__ == "__main__":
    lst = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
    targets = [23, 50]
    for target in targets:
        idx = Binary_search(lst, target)
        print(f"Target {target} {'found at index ' + str(idx) if idx != -1 else 'not found'}")
```

**Minesweeper Python Project**

```python
import random

def create_board(r, c, m):
    b = [[' '] * c for _ in range(r)]
    placed = 0
    while placed < m:
        row, col = random.randint(0, r - 1), random.randint(0, c - 1)
        if b[row][col] != 'M': b[row][col] = 'M'; placed += 1
    return b

def adjacent_mines(b, r, c):
    count = 0
    for i in range(max(0, r - 1), min(len(b), r + 2)):
        for j in range(max(0, c - 1), min(len(b[0]), c + 2)):
            if b[i][j] == 'M': count += 1
    return count

def show_board(pb):
    for row in pb: print(' | '.join(row)); print('-' * (len(row) * 4 - 1))

def play(rows=9, cols=9, mines=10):
    b = create_board(rows, cols, mines)
    pb = [['*'] * cols for _ in range(rows)]
    ml = mines
    while True:
        show_board(pb); print(f"Mines left: {ml}")
        try:
            r = int(input(f"Row (0-{rows - 1}): ")); c = int(input(f"Col (0-{cols - 1}): "))
            if not (0 <= r < rows and 0 <= c < cols): print("Invalid input.")
            elif b[r][c] == 'M': show_board(b); print("BOOM! Game Over!"); break
            else:
                adj = adjacent_mines(b, r, c); pb[r][c] = str(adj)
                if all(b[i][j] == 'M' or pb[i][j] != '*' for i in range(rows) for j in range(cols)):
                    show_board(b); print("You win!"); break
        except ValueError: print("Invalid input.")

if __name__ == "__main__":
    play()
```

**Sudoku Solver Python**

```python
def solve_sudoku(b):
    def v(r, c, n):
        for i in range(9):
            if b[r][i] == n or b[i][c] == n or b[3 * (r // 3) + i // 3][3 * (c // 3) + i % 3] == n:
                return False
        return True

    def s():
        for r in range(9):
            for c in range(9):
                if b[r][c] == '.':
                    for n in map(str, range(1, 10)):
                        if v(r, c, n):
                            b[r][c] = n
                            if s():
                                return True
                            b[r][c] = '.'
                    return False
        return True

    return s()
```

```python
import random

def make_model(lyrics, order=2):
    """Generates a Markov model from lyrics."""
    model = {}
    for i in range(len(lyrics) - order):
        state = tuple(lyrics[i:i + order])
        next_word = lyrics[i + order]
        model.setdefault(state, {}).setdefault(next_word, 0)
        model[state][next_word] += 1
    return model

def make_text(model, length=100, seed=None):
    """Generates text based on the Markov model."""
    seed = seed or random.choice(list(model))
    text = list(seed)
    for _ in range(length):
        current_state = tuple(text[-len(seed):])
        if current_state in model:
            choices, weights = zip(*model[current_state].items())
            next_word = random.choices(choices, weights=weights)[0]
            text.append(next_word)

    return ' '.join(text)

lyrics = ["I", "love", "to", "code", "in", "Python", "and", "R", "It's", "so", "much", "fun"]
model = make_model(lyrics, order=2)
generated_text = make_text(model, length=20)
print(generated_text)
```

**Pong Python Project**
In this Christian Thompson tutorial, you will learn how to recreate the classic pong game in Python. You will be working with the os and turtle Python modules which are great for creating graphics for games.

```python
import turtle

# Window setup
wn = turtle.Screen()
wn.title("Pong by OpenAI")  # Changed title
wn.bgcolor("black")
wn.setup(width=800, height=600)
wn.tracer(0)  # Stops the window from updating, speeds up the game

# Score
score_a = 0
score_b = 0

# Paddle A
paddle_a = turtle.Turtle()
paddle_a.speed(0)  # Animation speed, 0 is the fastest
paddle_a.shape("square")
paddle_a.color("white")
paddle_a.shapesize(stretch_wid=5, stretch_len=1)
paddle_a.penup()  # Don't draw lines when moving
paddle_a.goto(-350, 0)

# Paddle B
paddle_b = turtle.Turtle()
paddle_b.speed(0)
paddle_b.shape("square")
paddle_b.color("white")
paddle_b.shapesize(stretch_wid=5, stretch_len=1)
paddle_b.penup()
paddle_b.goto(350, 0)

# Ball
ball = turtle.Turtle()
ball.speed(0)
ball.shape("square")
ball.color("white")
ball.penup()
ball.goto(0, 0)
ball.dx = 0.1  # Ball's horizontal movement speed
ball.dy = -0.1  # Ball's vertical movement speed

# Pen for score display
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 260)
pen.write("Player A: 0  Player B: 0", align="center", font=("Courier", 24, "normal"))

# Functions for paddle movement
def paddle_a_up():
    y = paddle_a.ycor()  # Get y coordinate
    y += 20
    paddle_a.sety(y)

def paddle_a_down():
    y = paddle_a.ycor()
    y -= 20
    paddle_a.sety(y)

def paddle_b_up():
    y = paddle_b.ycor()
    y += 20
    paddle_b.sety(y)

def paddle_b_down():
    y = paddle_b.ycor()
    y -= 20
    paddle_b.sety(y)

# Keyboard bindings
wn.listen()  # Listen for keyboard input
wn.onkeypress(paddle_a_up, "w")
wn.onkeypress(paddle_a_down, "s")
wn.onkeypress(paddle_b_up, "Up")
wn.onkeypress(paddle_b_down, "Down")

# Main game loop
while True:
    wn.update()  # Updates the screen

    # Move the ball
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)

    # Border checking
    if ball.ycor() > 290:
        ball.sety(290)
        ball.dy *= -1  # Reverse direction

    if ball.ycor() < -290:
        ball.sety(-290)
        ball.dy *= -1

    if ball.xcor() > 390:
        ball.goto(0, 0)
        ball.dx *= -1
        score_a += 1
        pen.clear()  # Clear previous score
        pen.write(f"Player A: {score_a}  Player B: {score_b}", align="center", font=("Courier", 24, "normal"))

    if ball.xcor() < -390:
        ball.goto(0, 0)
        ball.dx *= -1
        score_b += 1
        pen.clear()
        pen.write(f"Player A: {score_a}  Player B: {score_b}", align="center", font=("Courier", 24, "normal"))

    # Paddle and ball collisions
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() + 40 and ball.ycor() > paddle_b.ycor() - 40):
        ball.setx(340)
        ball.dx *= -1

    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() + 40 and ball.ycor() > paddle_a.ycor() - 40):
        ball.setx(-340)
        ball.dx *= -1
```

**Snake Python Project**    Run the code in the local system

```python
import time
import random
import sys

def make_grid(w, h, s, f):
    g = [[' ' for _ in range(w)] for _ in range(h)]
    for r, c in s:
        if 0 <= r < h and 0 <= c < w:
            g[r][c] = '#'
    if 0 <= f[0] < h and 0 <= f[1] < w:
        g[f[0]][f[1]] = '@'
    return ["".join(row) for row in g]

def show_grid(g):
    for row in g:
        print(row)

def new_food(w, h, s):
    while True:
        food = (random.randint(0, h - 1), random.randint(0, w - 1))
        if food not in s:
            return food

def move(s, d):
    hx, hy = s[0]
    if d == 'w': nh = (hx - 1, hy)
    elif d == 's': nh = (hx + 1, hy)
    elif d == 'a': nh = (hx, hy - 1)
    elif d == 'd': nh = (hx, hy + 1)
    return [nh] + s[:-1]

def check_hit(s, w, h):
    hx, hy = s[0]
    if not (0 <= hx < h and 0 <= hy < w): return True
    if (hx, hy) in s[1:]: return True
    return False

def main():
    w, h = 20, 10
    snake = [(h // 2, w // 2)]
    food = new_food(w, h, snake)
    direc = 'd'
    score = 0

    while True:
        grid = make_grid(w, h, snake, food)
        show_grid(grid)
        print(f"Score: {score}")

        new_direc = input("Dir (w/a/s/d), q to quit: ").lower()
        if new_direc == 'q': break
        if new_direc in 'wasd': direc = new_direc

        snake = move(snake, direc)

        if snake[0] == food:
            snake.append(snake[-1])
            food = new_food(w, h, snake)
            score += 1

        if check_hit(snake, w, h):
            print("Game Over!")
            break

        time.sleep(0.2)

if __name__ == "__main__":
    main()
```
